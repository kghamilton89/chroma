```python
from unstructured.partition.html import partition_html
```

```python
cnn_lite_url = "https://lite.cnn.com/"
```

```python
elements = partition_html(url=cnn_lite_url)
```

```python
links = []
for element in elements:
    print(f"Checking element: {element}")

    if element.metadata.links is not None:
        print(f"Links found: {element.metadata.links}")
        relative_link = element.metadata.links[0]["url"][1:]
        print(f"Relative link: {relative_link}")

        if relative_link.startswith("2024"):
            final_link = f"{cnn_lite_url}{relative_link}"
            print(f"Appending link: {final_link}")
            links.append(final_link)
```
